```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

```python
def display_multiplication_table(number, limit=10):
    """Displays the multiplication table for a given number.

    Args:
        number: The integer for which to display the multiplication table.
        limit: The upper limit for the multiplication (default is 10).
    """
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
display_multiplication_table(7)
print("\n") # Add a newline for better readability between examples
display_multiplication_table(12, limit=5)
```

```python
def check_odd_even(number):
    """Checks if a given number is odd or even.

    Args:
        number: The integer to check.

    Returns:
        A string indicating whether the number is 'Even' or 'Odd'.
    """
    if number % 2 == 0:
        return f"{number} is Even"
    else:
        return f"{number} is Odd"

# Example usage:
print(check_odd_even(4))
print(check_odd_even(7))
print(check_odd_even(0))
print(check_odd_even(-3))
```

```python
def is_palindrome(number):
    """Checks if a given number is a palindrome.

    A number is a palindrome if it reads the same forwards and backward.

    Args:
        number: The integer to check.

    Returns:
        A string indicating whether the number is a palindrome or not.
    """
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage:
print(is_palindrome(121))
print(is_palindrome(12321))
print(is_palindrome(12345))
print(is_palindrome(1))
print(is_palindrome(10))
print(is_palindrome(0))
```

```python
knowledge_base = []
print("Created an empty list called 'knowledge_base' to store decision rules.")
```

```python
knowledge_base = [
    {
        'conditions': [
            {'fact': 'temperature', 'operator': '>', 'value': 25},
            {'fact': 'preference', 'operator': 'in', 'value': ['refreshing', 'cold']}
        ],
        'outcome': {'drink': 'iced tea'}
    },
    {
        'conditions': [
            {'fact': 'temperature', 'operator': '<=', 'value': 15},
            {'fact': 'preference', 'operator': 'in', 'value': ['hot', 'warm']}
        ],
        'outcome': {'drink': 'hot coffee'}
    },
    {
        'conditions': [
            {'fact': 'temperature', 'operator': '>', 'value': 15},
            {'fact': 'temperature', 'operator': '<=', 'value': 25},
            {'fact': 'preference', 'operator': 'in', 'value': ['neutral', 'any']}
        ],
        'outcome': {'drink': 'water'}
    },
    {
        'conditions': [
            {'fact': 'temperature', 'operator': '>', 'value': 20},
            {'fact': 'preference', 'operator': 'in', 'value': ['sweet', 'fruity']}
        ],
        'outcome': {'drink': 'lemonade'}
    }
]

print(f"Knowledge base defined with {len(knowledge_base)} rules.")
```

```python
def infer_drink_recommendation(facts, knowledge_base):
    """Infers a drink recommendation based on input facts and a knowledge base of rules.

    Args:
        facts (dict): A dictionary of current conditions (e.g., {'temperature': 28, 'preference': 'refreshing'}).
        knowledge_base (list): A list of rule dictionaries, each with 'conditions' and an 'outcome'.

    Returns:
        dict: The outcome of the first matched rule, or a default message if no rule matches.
    """
    for rule in knowledge_base:
        all_conditions_met = True
        for condition in rule['conditions']:
            fact_name = condition['fact']
            operator = condition['operator']
            value = condition['value']

            if fact_name not in facts:
                all_conditions_met = False
                break

            current_fact_value = facts[fact_name]

            if operator == '>':
                if not (current_fact_value > value):
                    all_conditions_met = False
                    break
            elif operator == '<=':
                if not (current_fact_value <= value):
                    all_conditions_met = False
                    break
            elif operator == 'in':
                # Ensure current_fact_value is iterable for 'in' operator, or handle specific cases
                if not (current_fact_value in value):
                    all_conditions_met = False
                    break
            elif operator == '==':
                if not (current_fact_value == value):
                    all_conditions_met = False
                    break
            # Add more operators as needed (e.g., '>=', '<', '!=')

        if all_conditions_met:
            return rule['outcome']

    return {'drink': 'No specific recommendation'}

print("Defined the 'infer_drink_recommendation' function.")
```

```python
print("Demonstrating the expert system with various scenarios:")

# Scenario 1: Hot day, refreshing preference
facts1 = {'temperature': 28, 'preference': 'refreshing'}
recommendation1 = infer_drink_recommendation(facts1, knowledge_base)
print(f"\nFacts: {facts1} -> Recommendation: {recommendation1['drink']}")

# Scenario 2: Cold day, warm preference
facts2 = {'temperature': 10, 'preference': 'hot'}
recommendation2 = infer_drink_recommendation(facts2, knowledge_base)
print(f"Facts: {facts2} -> Recommendation: {recommendation2['drink']}")

# Scenario 3: Moderate day, neutral preference
facts3 = {'temperature': 20, 'preference': 'neutral'}
recommendation3 = infer_drink_recommendation(facts3, knowledge_base)
print(f"Facts: {facts3} -> Recommendation: {recommendation3['drink']}")

# Scenario 4: Hot day, sweet preference
facts4 = {'temperature': 26, 'preference': 'sweet'}
recommendation4 = infer_drink_recommendation(facts4, knowledge_base)
print(f"Facts: {facts4} -> Recommendation: {recommendation4['drink']}")

# Scenario 5: Another moderate day, different neutral preference
facts5 = {'temperature': 18, 'preference': 'any'}
recommendation5 = infer_drink_recommendation(facts5, knowledge_base)
print(f"Facts: {facts5} -> Recommendation: {recommendation5['drink']}")

# Scenario 6: No specific rule applies (e.g., very cold, refreshing preference)
facts6 = {'temperature': 5, 'preference': 'refreshing'}
recommendation6 = infer_drink_recommendation(facts6, knowledge_base)
print(f"Facts: {facts6} -> Recommendation: {recommendation6['drink']}")

# Scenario 7: No specific rule applies (e.g., hot, warm preference)
facts7 = {'temperature': 30, 'preference': 'warm'}
recommendation7 = infer_drink_recommendation(facts7, knowledge_base)
print(f"Facts: {facts7} -> Recommendation: {recommendation7['drink']}")
```
