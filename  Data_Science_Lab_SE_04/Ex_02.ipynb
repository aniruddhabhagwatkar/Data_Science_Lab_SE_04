## Breadth-First Search (BFS)

BFS is an algorithm for traversing or searching tree or graph data structures. It starts at the tree root (or some arbitrary node of a graph, sometimes referred to as a 'search key') and explores all of the neighbor nodes at the present depth prior to moving on to the nodes at the next depth level.

```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    visited.add(start_node)

    bfs_traversal = []

    while queue:
        current_node = queue.popleft()
        bfs_traversal.append(current_node)

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return bfs_traversal

# Example Graph (Adjacency List Representation)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("BFS Traversal starting from node 'A':")
print(bfs(graph, 'A'))

print("\nBFS Traversal starting from node 'C':")
print(bfs(graph, 'C'))
```

## Minimax Algorithm

Minimax is a decision rule used in artificial intelligence, decision theory, game theory, statistics, and philosophy for minimizing the possible loss for a worst-case (maximum loss) scenario. When dealing with gains, it is termed 'maximin'—to maximize the minimum gain. It's often used in two-player turn-based games like Tic-Tac-Toe, Chess, or Checkers.

The algorithm works by recursively exploring the game tree, assigning a score to each possible move. The 'maximizer' player aims to choose the move that leads to the highest score, while the 'minimizer' player aims to choose the move that leads to the lowest score (from the maximizer's perspective). The recursion continues until a terminal state (end of the game) is reached, or a predefined depth limit is met.

```python
def minimax(node, depth, maximizing_player, game_tree):
    # If node is a leaf node (terminal state) or depth limit is reached
    if node not in game_tree or depth == 0:
        return node, game_tree.get(node, node) # Return node and its utility value

    if maximizing_player:
        max_eval = -float('inf')
        best_move = None
        for child in game_tree[node]:
            _, eval = minimax(child, depth - 1, False, game_tree)
            if eval > max_eval:
                max_eval = eval
                best_move = child
        return best_move, max_eval
    else:
        min_eval = float('inf')
        best_move = None
        for child in game_tree[node]:
            _, eval = minimax(child, depth - 1, True, game_tree)
            if eval < min_eval:
                min_eval = eval
                best_move = child
        return best_move, min_eval

# Example Game Tree (represented as a dictionary)
# Leaf nodes have integer values (utility values)
# Internal nodes have lists of their children
game_tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [3, 5], # Leaf nodes directly give utility
    'E': [2, 9],
    'F': [1, 0],
    'G': [4, 8]
}

# To make the leaf nodes consistent with the recursive calls,
# we'll represent them as single-element lists that are not present as keys in game_tree
# This allows the `if node not in game_tree` check to work.
# Alternatively, we could modify the `minimax` function to handle integers directly for leaf nodes.
# For this example, let's represent the leaves as if they were children of a non-existent node
# that return their value directly.

# A more typical representation for this type of problem might be:
# The keys are parent nodes, and values are lists of children nodes.
# If a node is not a key, it's a leaf node whose value can be directly retrieved.

# Let's adjust the game_tree for a clearer example with terminal values

# In this revised representation, if a node is not a key in the dictionary,
# it's considered a terminal node with its value as the utility.
revised_game_tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [3, 12],
    'E': [8, 2],
    'F': [4, 6],
    'G': [14, 5]
}

# For this implementation to work, we need to ensure that leaf nodes (the integer values)
# are not keys in the game_tree, and the minimax function can directly return their value.
# Let's create a tree where the actual utility values are directly at the end of the branches
# And the children are strings that represent intermediate nodes or the final utility itself.

# A better way to define the game tree for this minimax implementation:
# Keys are non-terminal nodes, values are lists of child nodes.
# Terminal nodes are not keys and have an associated utility value.

minimax_game_tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H', 'I'],
    'E': ['J', 'K'],
    'F': ['L', 'M'],
    'G': ['N', 'O']
}

# Utility values for terminal nodes:
utility_values = {
    'H': 3,
    'I': 5,
    'J': 2,
    'K': 9,
    'L': 1,
    'M': 0,
    'N': 4,
    'O': 8
}

# We need to pass the utility_values to the minimax function.
# Let's modify the minimax function slightly to handle this structure.

def minimax_with_utility(node, depth, maximizing_player, game_tree, utilities):
    # If node is a leaf node (terminal state) or depth limit is reached
    # A node is terminal if it's not in the game_tree dictionary (i.e., it has no children)
    # and it has a utility value defined.
    if node in utilities or depth == 0:
        return node, utilities.get(node, 0) # Return node and its utility value, 0 if not found

    if maximizing_player:
        max_eval = -float('inf')
        best_move = None
        for child in game_tree.get(node, []): # Get children, or empty list if no children
            _, eval = minimax_with_utility(child, depth - 1, False, game_tree, utilities)
            if eval > max_eval:
                max_eval = eval
                best_move = child
        return best_move, max_eval
    else:
        min_eval = float('inf')
        best_move = None
        for child in game_tree.get(node, []): # Get children, or empty list if no children
            _, eval = minimax_with_utility(child, depth - 1, True, game_tree, utilities)
            if eval < min_eval:
                min_eval = eval
                best_move = child
        return best_move, min_eval


print("Minimax Example:")
start_node = 'A'
depth = 3 # Max depth to search

best_move, optimal_value = minimax_with_utility(start_node, depth, True, minimax_game_tree, utility_values)

print(f"For the maximizing player starting at node '{start_node}' with a search depth of {depth}:")
print(f"Optimal move: {best_move}")
print(f"Optimal value: {optimal_value}")

print("\nLet's consider a scenario where the 'min' player starts at node 'B'")
best_move_B, optimal_value_B = minimax_with_utility('B', 2, False, minimax_game_tree, utility_values)
print(f"For the minimizing player starting at node 'B' with a search depth of 2:")
print(f"Optimal move: {best_move_B}")
print(f"Optimal value: {optimal_value_B}")
```

## A* Search Algorithm

A* is a popular pathfinding algorithm that finds the shortest path between a starting and a target node in a graph. It uses a heuristic function to estimate the cost from the current node to the goal, combining this with the actual cost from the start node to the current node. This makes it an informed search algorithm, often more efficient than uninformed search algorithms like BFS or DFS.

```python
import heapq

def a_star_search(graph, start, goal, heuristic):
    # The set of discovered nodes that may need to be (re-)expanded.
    # Initially, only the start node is known.
    # This is a min-heap, so elements are (f_score, node)
    open_set = [(heuristic[start], start)]

    # For node n, came_from[n] is the node immediately preceding it on the cheapest path from start
    # to n currently known.
    came_from = {}

    # For node n, g_score[n] is the cost of the cheapest path from start to n currently known.
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0

    # For node n, f_score[n] = g_score[n] + heuristic[n]. f_score[n] represents our current best guess as to
    # how cheap a path from start to goal can be if it goes through n.
    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic[start]

    while open_set:
        # current is the node in open_set with the lowest f_score value
        current_f, current = heapq.heappop(open_set)

        if current == goal:
            return reconstruct_path(came_from, current)

        for neighbor, weight in graph[current].items():
            tentative_g_score = g_score[current] + weight

            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = tentative_g_score + heuristic[neighbor]
                heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return None  # Path not found

def reconstruct_path(came_from, current):
    total_path = [current]
    while current in came_from:
        current = came_from[current]
        total_path.append(current)
    return total_path[::-1]

# Example Graph (Adjacency List with weights)
graph_a_star = {
    'A': {'B': 1, 'C': 3},
    'B': {'A': 1, 'D': 3, 'E': 6},
    'C': {'A': 3, 'F': 2},
    'D': {'B': 3},
    'E': {'B': 6, 'F': 1},
    'F': {'C': 2, 'E': 1}
}

# Heuristic function (straight-line distance to goal 'F')
# In a real scenario, this would be calculated based on coordinates or domain knowledge.
heuristic_a_star = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 8,
    'E': 1,
    'F': 0
}

start_node = 'A'
goal_node = 'F'

path = a_star_search(graph_a_star, start_node, goal_node, heuristic_a_star)

if path:
    print(f"A* Path from {start_node} to {goal_node}: {path}")
else:
    print(f"No path found from {start_node} to {goal_node}")

start_node_2 = 'C'
goal_node_2 = 'D'

path_2 = a_star_search(graph_a_star, start_node_2, goal_node_2, heuristic_a_star)

if path_2:
    print(f"\nA* Path from {start_node_2} to {goal_node_2}: {path_2}")
else:
    print(f"\nNo path found from {start_node_2} to {goal_node_2}")
```

## Depth-First Search (DFS)

DFS is an algorithm for traversing or searching tree or graph data structures. The algorithm starts at the root node (selecting some arbitrary node as the root node in the case of a graph) and explores as far as possible along each branch before backtracking.

```python
def dfs(graph, start_node, visited=None):
    if visited is None:
        visited = set()

    dfs_traversal = []
    stack = [start_node]

    while stack:
        current_node = stack.pop()
        if current_node not in visited:
            visited.add(current_node)
            dfs_traversal.append(current_node)

            # Explore neighbors. For consistent output, sort neighbors and reverse
            # so that smaller/earlier nodes are processed first (pop from stack).
            for neighbor in sorted(graph[current_node], reverse=True):
                if neighbor not in visited:
                    stack.append(neighbor)

    return dfs_traversal

# Recursive DFS implementation (alternative)
def dfs_recursive(graph, start_node, visited=None, traversal=None):
    if visited is None:
        visited = set()
    if traversal is None:
        traversal = []

    visited.add(start_node)
    traversal.append(start_node)

    for neighbor in sorted(graph[start_node]): # Sorting for consistent output
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited, traversal)

    return traversal

# Example Graph (same as BFS example)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("DFS Traversal (iterative) starting from node 'A':")
print(dfs(graph, 'A'))

print("\nDFS Traversal (recursive) starting from node 'A':")
# Need to re-initialize visited and traversal for each call if not handled inside
print(dfs_recursive(graph, 'A', visited=set(), traversal=[]))

print("\nDFS Traversal (iterative) starting from node 'F':")
print(dfs(graph, 'F'))
```
