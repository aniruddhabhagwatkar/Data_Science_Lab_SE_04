```python
import numpy as np

# Generate even numbers from 1 to 100
even_numbers = np.arange(2, 101, 2)

print("Dataset of even numbers (1-100):")
print(even_numbers)
print(f"Total numbers generated: {len(even_numbers)}")
```

```python
# Calculate descriptive statistics for the 'even_numbers' dataset

# Mean
mean_even_numbers = np.mean(even_numbers)
print(f"Mean of even_numbers: {mean_even_numbers:.2f}")

# Median
median_even_numbers = np.median(even_numbers)
print(f"Median of even_numbers: {median_even_numbers:.2f}")

# Variance
variance_even_numbers = np.var(even_numbers)
print(f"Variance of even_numbers: {variance_even_numbers:.2f}")

# Standard Deviation
std_dev_even_numbers = np.std(even_numbers)
print(f"Standard Deviation of even_numbers: {std_dev_even_numbers:.2f}")
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(even_numbers, fill=True, color='blue', label='Distribution of Even Numbers')
plt.axvline(x=mean_even_numbers, color='red', linestyle='--', label=f'Mean: {mean_even_numbers:.2f}')
plt.title('Distribution of Even Numbers with Mean Value (KDE Plot)')
plt.xlabel('Even Number Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a standard normal distribution (mean=0, standard deviation=1)
# np.random.randn() is a convenient function for generating samples from a standard normal distribution
standard_normal_samples = np.random.randn(1000)

print(f"Generated {len(standard_normal_samples)} samples from a standard normal distribution.")
print(f"Sample Mean: {np.mean(standard_normal_samples):.2f}")
print(f"Sample Standard Deviation: {np.std(standard_normal_samples):.2f}")

# Plotting the histogram of the generated samples
plt.figure(figsize=(10, 6))
plt.hist(standard_normal_samples, bins=30, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
plt.title('Histogram of Standard Normal Distribution Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(standard_normal_samples, fill=True, color='purple', label='Standard Normal Distribution')
plt.title('KDE Plot of Standard Normal Distribution Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Simulate a coin toss experiment
num_tosses = 1000  # Number of coin tosses
results = np.random.choice(['Heads', 'Tails'], size=num_tosses)

# Count occurrences of Heads and Tails
heads_count = np.sum(results == 'Heads')
tails_count = np.sum(results == 'Tails')

# Compute probabilities
prob_heads = heads_count / num_tosses
prob_tails = tails_count / num_tosses

print(f"Total coin tosses: {num_tosses}")
print(f"Number of Heads: {heads_count} (Probability: {prob_heads:.2f})")
print(f"Number of Tails: {tails_count} (Probability: {prob_tails:.2f})")
```
